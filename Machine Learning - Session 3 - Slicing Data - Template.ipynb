```python
# Set the working directory
import os
os.chdir("/Users/skywalker/Downloads")

#Read the cars.csv data
import pandas as pd
cars = pd.read_csv("cars.csv")
```

#### 1. Choosing multiple columns

```python
#Select the coumns MPG, Weight and Horsepower
cars[["MPG","Weight","Horsepower"]].head()
```

```python
#OR
#cars[[1,5,4]].head() #not working
```

```python
#Using iloc (iloc allows you to give the indices)
cars.iloc[:,[1,5,4]].head()
```

```python
#OR Using loc (loc allows you to give the names)
cars.loc[:,["MPG","Weight","Horsepower"]].head()
```

#### 3. Choosing rows

```python
#A. Choosing a single row from a single variable
cars.MPG[1]
```

```python
#OR
cars["MPG"][1]
```

```python
#cars[1]["MPG"] doesnt work
```

```python
#OR using iloc
cars.iloc[1,1]
```

```python
#OR using loc
cars.loc[1,"MPG"]
```

```python
#B. Slicing few rows from a single variable
cars.MPG[10:20]
```

```python
#OR using loc
cars.loc[10:20,"MPG"] #gives 20th row as well even with iloc we'll get till 19th
```

```python
#C. Choosing some specific rows from  particular variable
cars.loc[[1,4,8],"MPG"]
```

```python
#D. Choosing some observations from all the variables
cars.loc[[1,4,8],:]
```

```python
#OR simply
cars.loc[[1,4,6]]
```

#### 5. Condtional Subsetting

```python
#a) Extract all the information of the cars which has MPG more than 40
for i in range(0,len(cars)):
    if int(cars.MPG[i])>40:
        print(cars.loc[i,:])
```

```python
#b) Extract the information of the car/s which has only 3 cylinders
cars[cars.MPG>40]
```

```python
#c) Extract the information of all the cars originated in Europe and has Acceleration more than 20
cars[(cars["Origin"]=="Europe") & (cars.Acceleration>20)] #() and & are very important
```
